```python
import pandas as pd
```

```python
df = pd.read_csv(r"dataset/Sign_Language_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna()
df_clean
```

```python
df_clean.isnull().sum()
```

```python
def normalize_row(row):
    x1, y1 = row['X1'], row['Y1']
    for i in range(1, 22):
        row[f'X{i}'] -= x1
        row[f'Y{i}'] -= y1
    return row

df_clean = df_clean.apply(normalize_row, axis=1)
```

```python
df_clean
```

```python
import matplotlib.pyplot as plt

row = df_clean.iloc[0]  # مثال على أول صورة
x = [row[f'X{i}'] for i in range(1,22)]
y = [row[f'Y{i}'] for i in range(1,22)]

plt.scatter(x, y)
plt.gca().invert_yaxis()
plt.title(f"Hand Landmarks for {row['Letter']}")
plt.show()
```

```python
letters = df_clean['Letter'].unique()

letters
```

```python
len(letters)
```

```python
for letter in letters:
    subset = df_clean[df_clean['Letter'] == letter].iloc[0]
    print(subset)
```

```python
for letter in letters:
    subset = df_clean[df_clean['Letter'] == letter]
    print(subset)
```

```python
for letter in letters:
    subset = df_clean[df_clean['Letter'] == letter].iloc[0]
    avg_y = [subset[f'Y{i}'] for i in range(1, 22)]
    avg_x = [subset[f'X{i}'] for i in range(1, 22)]

    plt.scatter(avg_x, avg_y, c='red')
    plt.plot(avg_x, avg_y, c='blue')
    plt.gca().invert_yaxis()
    plt.title(f"Average landmarks for {letter}")
    plt.show()
```

```python
for letter in letters:
    subset = df_clean[df_clean['Letter'] == letter]
    avg_y = [subset[f'Y{i}'].mean() for i in range(1, 22)]
    avg_x = [subset[f'X{i}'].mean() for i in range(1, 22)]

    plt.scatter(avg_x, avg_y, c='red')
    plt.plot(avg_x, avg_y, c='blue')
    plt.gca().invert_yaxis()
    plt.title(f"Average landmarks for {letter}")
    plt.show()
```
